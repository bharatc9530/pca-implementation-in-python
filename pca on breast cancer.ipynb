```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
breast = load_breast_cancer()
```

```python
y=breast.keys()
y
```

```python
breast_data = breast.data
```

```python
breast_data.shape
```

```python
breast_labels = breast.target
```

```python
breast_labels.shape
```

```python
labels = np.reshape(breast_labels,(569,1))
```

```python
final_breast_data = np.concatenate([breast_data,labels],axis=1)
```

```python
final_breast_data.shape
```

```python
final_breast_data.shape
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
features = breast.feature_names
```

```python
features
```

```python
features_labels = np.append(features,'label')
```

```python
breast_dataset.columns = features_labels
```

```python
breast_dataset.head(20)
```

```python
breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)
```

```python
breast_dataset.tail()
```

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) 
```

```python
x.shape
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
```

```python
normalised_breast = pd.DataFrame(x,columns=feat_cols)
```

```python
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=4)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2','principal component 3','principal component 4'])
```

```python
principal_breast_Df["principal component 1"]
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=30)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast)
```

```python
principal_breast_Df
```

```python
z=(x-principal_breast_Df).mean()
```

```python
z
```

```python
from numpy import linalg as LA
max_comp=30
start=1
error_record=[]
for i in range(start,max_comp):
    pca = PCA(n_components=i, random_state=42)
    pca2_results = pca.fit_transform(x)
    pca2_proj_back=pca.inverse_transform(pca2_results)
    total_loss=LA.norm((x-pca2_proj_back),None)
    error_record.append(total_loss)

plt.clf()
plt.figure(figsize=(10,10))
plt.title("reconstruct error of pca")
plt.plot(error_record,'r')
plt.xticks(range(len(error_record)), range(start,max_comp), rotation='vertical')
plt.xlim([-1, len(error_record)])
plt.show()
```
